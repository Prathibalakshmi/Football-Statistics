## Reading the data

```python
import pandas as pd
import numpy as np
fb=pd.read_csv(r"C:\Users\HP\Downloads\Laliga.csv")
fb.head()
```

```python
fb.shape
```

```python
fb.isnull().sum()
```

```python
# Replace dashes with zeros in the entire DataFrame
fb = fb.replace('-', 0)
fb.head()
```

```python
# Define column names for better accessibility
column_names = [
    'Pos', 'Team', 'Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn',
    'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion', 'Runner-up', 'Third',
    'Fourth', 'Fifth', 'Sixth', 'T', 'Debut', 'Since/LastApp', 'BestPosition'
]

# Assign column names to the DataFrame
fb.columns = column_names
fb.head()
```

```python
import copy
fb2=copy.deepcopy(fb)
fb2.head()
```

```python
# Rename the columns for better accessibility
fb2.columns = fb2.iloc[0]  # Set the first row as column names
fb2= fb2[1:]  # Remove the first row (it contains column descriptions)
fb2.head()
```

```python
## Print all the teams which have started playing between 1930-1980.

# Convert 'Debut' column to extract the starting year
fb2['Debut_Year']=fb2['Debut'].str.extract(r'(\d{4})')

# Convert 'Debut_Year' column to numeric (considering some years may have strings)
fb2['Debut_Year']=pd.to_numeric(fb2['Debut_Year'],errors='coerce')

fb2.head()
```

```python
# Filter teams that started playing between 1930 and 1980
filtered_teams=fb2[(fb2['Debut_Year']>=1930) & (fb2['Debut_Year']<=1980)]
filtered_teams.head()
```

```python
# Print the names of teams that started playing between 1930 and 1980
print("Teams that started playing between 1930 and 1980:")
print(filtered_teams["Team"].tolist())
```

```python
# Print the list of teams which came Top 5 in terms of points
fb2['Points'] = pd.to_numeric(fb2['Points'], errors='coerce')

# Get the top 5 teams based on points
top_teams = fb2.nlargest(5, 'Points')
print("Teams that came top 5 in Points:")
print(top_teams['Team'].tolist())
```

```python
import pandas as pd

def Goal_diff_count(data):
    # Calculate goal differences
    data['GoalDiff'] = data['GoalsFor'].astype(int) - data['GoalsAgainst'].astype(int)
    
    # Create a dictionary to store team names and their goal differences
    goal_diff_dict = dict(zip(data['Team'], data['GoalDiff']))
    
    # Find team with maximum goal difference
    max_goal_diff_team = data.loc[data['GoalDiff'].idxmax()]
    
    # Find team with minimum goal difference
    min_goal_diff_team = data.loc[data['GoalDiff'].idxmin()]
    
    return goal_diff_dict, max_goal_diff_team, min_goal_diff_team

# Call the function Goal_diff_count with the DataFrame 
goal_diff_dict, max_goal_diff_team, min_goal_diff_team = Goal_diff_count(fb2)

# Print all teams with their goal differences
print("Teams with their Goal Differences:")
for team, goal_diff in goal_diff_dict.items():
    print(f"{team}: {goal_diff}")

# Print team with maximum goal difference
print("\nTeam with Maximum Goal Difference:")
print(max_goal_diff_team['Team'], "with a goal difference of", max_goal_diff_team['GoalDiff'])

# Print team with minimum goal difference
print("\nTeam with Minimum Goal Difference:")
print(min_goal_diff_team['Team'], "with a goal difference of", min_goal_diff_team['GoalDiff'])
```

```python
#Create a new column with name “Winning Percent” and append it to the data set (5 points)

#Percentage of Winning = (GamesWon / GamesPlayed)*100

#If there are any numerical error, replace it with 0%

#Print the top 5 teams which has the highest Winning percentage

# Convert 'GamesWon' and 'GamesPlayed' columns to numeric

fb2['GamesWon']= pd.to_numeric(fb2['GamesWon'],errors='coerce')
fb2['GamesPlayed']= pd.to_numeric(fb2['GamesPlayed'],errors='coerce')

# Calculate Winning Percent and handle possible numerical errors

fb2['Winning Percent']=(fb2['GamesWon']/ fb2['GamesPlayed'])*100
fb2['Winning Percent'].fillna(0,inplace=True)

# Sort the DataFrame by 'Winning Percent' in descending order and print top 5 teams
top_5_teams= fb2.nlargest(5,'Winning Percent')
print("Top 5 teams with the highest Winning Percentage:")
print(top_5_teams[['Team','Winning Percent']])
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions (10 points)

#Eg: Best Position     Points

#         1                   25000

 #        2                    7000
    


# Convert 'Points' column to numeric
fb2['Points'] = pd.to_numeric(fb2['Points'], errors='coerce')

# Group teams based on their 'BestPosition' and calculate the sum of points for each position
sum_points_by_position = fb2.groupby('BestPosition')['Points'].sum().reset_index()

# Rename columns for better readability
sum_points_by_position.columns = ['Best Position', 'Total Points']

# Print the sum of points for each best position
print("Sum of points for each Best Position:")
print(sum_points_by_position)
```
